```python
#NUMBER 2(A)
import matplotlib.pyplot as plt
Breath_held=[22.22,30.57,17.47,22.39,26.90,36.85,27.33,29.55,13.87,34.66,60.75,67.41,42.19,59.74,52.64,43.37,73.27,59.09,51.15,58.32]
plt.figure(figsize=(11,6))
plt.hist(Breath_held, bins=10,color='darkblue', edgecolor='black')
plt.xlabel('Breathe Held')
plt.ylabel('Frequency')
plt.title('An Histogram of Breathe Held')
plt.show()
```

```python
#NUMBER 2(B)
import matplotlib.pyplot as plt
male_breath=[60.75,67.41,42.19,59.74,52.64,43.37,73.27,59.09,51.15,58.32]
female_breath=[22.22,30.57,17.47,22.39,26.90,36.85,27.33,29.55,13.87,34.66]
plt.plot(male_breath, [0] * len(male_breath), 'o', label='Male', color='pink')
plt.plot(female_breath, [1] * len(female_breath), 'o', label='Female', color='orange')
plt.title('Side by side plot of breath held by gender')
plt.xlabel('Breath held')
plt.yticks([0,1], ['Male', 'Female'])
plt.legend()
plt.show()
```

```python
#Number 3(A)
import matplotlib.pyplot as plt
with_music=[304,257,174,214,69,317,387,47,157,0,332,308,317,286,236,299,206,278,188,43,0,0,0,0,0]
without_music=[292,270,47,288,524,292,364,316,287,75,282,149,274,319,213,3,324,2,128,219,94,164,0,0,0]
plt.plot(with_music, [0] * len(with_music), 'o', color='purple', label='With Music')
plt.plot(without_music, [1] * len(without_music), 'o', color='red', label='Without Music')
plt.title('Growth with and without music')
plt.yticks([0,1], ['With music', 'Without music'])
plt.legend(loc='best')
plt.show()
```
